```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as so

c = 4
x = np.linspace(-np.pi/2, np.pi/2, 500)
x2 = np.where((x > np.pi/c) | (x < -np.pi/c))

storm = ((np.cos((x - np.pi/4)*2))**3)*2
x1 = np.where((storm < 0) | (x < 0))
storm[x1] = 0
storm[x<0]=-storm[x>0]

y = -0.4
a = np.arcsin(y)
had1 = (np.sin(c*x - a)) - y
had1[x2]=0

fig, ax = plt.subplots(3, 1)
ax[0].plot(x, storm, color='purple')
ax[0].axvline(0, color='black')
ax[0].axhline(0, color='black')
ax[0].set_title('Storm Track')

ax[1].plot(x, had1, color='purple')
ax[1].axvline(0, color='black')
ax[1].axhline(0, color='black')
ax[1].set_title('Hadley Cell')

ax[2].plot(x, storm+had1, color='purple')
ax[2].axvline(0, color='black')
ax[2].axhline(0, color='black')
ax[2].set_title('Idealized Heat Transport')

fig.tight_layout()
plt.show()
fig.savefig('/Users/ragen/AREDI/ACDC/ISCA/had_storm_idealized', format='png', bbox_inches='tight')
```
